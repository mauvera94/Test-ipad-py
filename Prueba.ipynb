```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Create a simple DataFrame with random data
data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],
    'Sales': np.random.randint(100, 500, 12),
    'Expenses': np.random.randint(50, 400, 12)
}

df = pd.DataFrame(data)

# Display the DataFrame
print(df)

# Basic statistics
print("\nBasic statistics:")
print(df.describe())

# Plot Sales and Expenses as a bar plot
df.plot(x='Month', y=['Sales', 'Expenses'], kind='bar', figsize=(10,6))
plt.title('Monthly Sales vs Expenses')
plt.ylabel('Amount')
plt.show()

# Line plot for Sales over months
df.plot(x='Month', y='Sales', kind='line', marker='o', figsize=(10,6))
plt.title('Monthly Sales Trend')
plt.ylabel('Sales Amount')
plt.show()
```
